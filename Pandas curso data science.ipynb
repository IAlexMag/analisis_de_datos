***Utilizando librería pandas***

Serie es un objeto de pandas diseñado para representar una estructura de una sola dimensión, los objetos de tip serie poseen dos arreglos uno para almacenar los valores y el segundo para alamcenar los índices

pd.series([1,2,3,4,5])

para poder asignar un nombre a la serie de datos y especificar el tipo de datos que se quiere almacenar se utiliza la siguiente sintaxis
pd.Series([1,2,3,4,5], index = ['a', 'b', 'c', 'd', 'e'], name = 'Numeros', dtype= str)

```
import pandas as pd # Importar librería pandas
import numpy as np # Importar librería numpy
a = pd.Series([1,2,3,4,5])
print(a.size)
print(a.dtype)
print(a.index)

# actualizar valores de la serie
a[0] = 0
print(a)

#para poder procionar los indices a una serie en pandas se utiliza la siguiente sintaxis

b = pd.Series([1,2,3,4,5], index=['a','b','c','d','e'])
print(b)

# actualizar valores mediante los indices
b['a'] = 0
print(b)
```

*COLUMNAS DEL DATA FRAME*

las columnas son igual a series

para poder agregar los valores correctos acorde a los indices es necesario especificar los indices en la serie acorde a los indices presentes en el dataframe.

```
diccionario = {
    'name': ['Juan', 'Pedro', 'Maria'],
    'age': [20, 30, 40],
    'gender': ['M', 'M', 'F']
}

data = pd.DataFrame(diccionario)
data.columns

calififcaciones = pd.Series(np.random.randint(0,10, 3))

#añadiendo una columna
data['calificaciones'] = calififcaciones
print(data)

#renombrar las columnas

data = data.rename(
    columns={'calificaciones': 'score'}
)
print(data)

#eliminar columnas

del data['score']
data
```

*LIMPIEZA DE DATOS*

En numpy y pandas la ausencia de datos e representa como NaN

para poder limpiar la data se tienen las siguientes maneras:
* Método dropna() que elimina las filas que contengan datos nulos
* data.fillna(valor a colocar) genera un nuevo df donde sustituye los valores nulos por las leyendas indicadas

* Se recomienda rellenar mediante un diccionario es decir: data.fillna({'name' : 'Sin nombre', 'email' : 'example@example.com'})

**Atributo loc** se usa para obtener filas con índices del tipo string
data.loc['a':'c'] con esta sntaxis se obtienen los valores de las columnas con indices del a al c
data.loc['a':'c', ['username', 'email']]

o bien
data.loc['a':'c'][['username', 'email']]

todo esto genera subdata frames


**ATRIBUTO ILOC**

**CONDICIONALES**
* Obtener todos los resultados que sean de dos o más goles
* Obtener el nombre de las selecciones que hayan disputado un partido en MADRID
* Obtener el promedio de goles de los equipos locales que hayan disputado partidos en Estados Unidos

* Obtener los 5 mayores resultados de selecciones con juegos disputados en Estados Unidos

```
df= df[df['country'] == 'United States'].sort_values('home_score', ascending=False).head(5)
```

**Búsqueda por rangos**

* Obtener todos los resultados de local entre 4 y 6 goles, con partidos disputados en Francia

```
df = df[df['home_team'].between(4, 6)]
```

**Búsqueda entre opciones**
* Obtener el nombre de las selecciones con home score mayor a 5 y con partidos disputados en Francia, Canada e Inglaterra

```
df = df[(df['home_score']>5) & (df['country'].isin(['France', 'Canada', 'England']))]
```

**Métodos de strings**

* startswith
* endswith
* contains

```
df = df[df['country'].str.contains('F')]
df = df[df['country'].str.endswith('e')]
df = df[df['country'].str.startswith('S')]
```

**Agrupamiento**
* Mostrar en consola la cantidad de partidos disputados por México
* Mostrar el país con más partidos disputados en Estados Unidos y Francia

Siempre que se quiera agrupar se hace mediante groupby lo que dar por resultados un objeto del tipo series groupby a las cuales se les pueden aplicar métodos de agregación.

```
df = df.groupby('country')['country'].count()
df = df[df['country'].isin(['United States', 'France'])].groupby('country')['country'].count().sort_values(ascending=False)
```
